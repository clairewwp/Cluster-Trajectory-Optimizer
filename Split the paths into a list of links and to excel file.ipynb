```python
for i in range(len(df['path'])):
    df['path'][i]=df['path'][i].strip('[]')
    df['path'][i]=df['path'][i].split(',')
```

```python
df_new=pd.DataFrame()
path_id=[]
links=[]
for i in range(len(df['path'])):##62
    for j in range(len(df['path'][i])):
        path_id.append(df['path id'][i])
        links.append(df['path'][i][j])
            
```

```python
df_new['path_id']=path_id
df_new['link']=links
```
